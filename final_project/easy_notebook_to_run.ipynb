```python
import pandas as pd
from final_project.notebooks.training_pipeline import auto_training_pipeline
```

```python
USER_REVIEWS_PATH = './data/video-games/processed/learning_user_reviews.csv'
SALES_PATH = './data/video-games/processed/learning_sales.csv'
COMBINED_PATH = './data/video-games/processed/learning_combined_video_games.csv'

user_reviews_df = pd.read_csv(USER_REVIEWS_PATH)
sales_df = pd.read_csv(SALES_PATH)
combined_df = pd.read_csv(COMBINED_PATH)
```

```python
user_reviews_df['user_review'] = user_reviews_df['user_review'].astype(str)
sales_df['user_review'] = sales_df['user_review'].astype(str)
combined_df['user_review'] = combined_df['user_review'].astype(str)
```

```python
perf_user_reviews, best_model_user_reviews, leaderboard_user_reviews = auto_training_pipeline(
    df=user_reviews_df,
    target_col='user_review',
    max_runtime_secs=300,
    max_models=10,
    sort_metric='mse',
    balance_classes=False,
    print_metrics=True,
    show_scoring_history=True,
    show_variable_importance=True,
    shutdown_h2o_after_train=False
)
```
